```python
import matplotlib.pyplot as plt
import pandas as pd
import japanize_matplotlib

df = pd.read_csv("/Users/poppokuri/Downloads/sample_pandas_6.csv")

category_df = pd.read_csv("/Users/poppokuri/Downloads/category.csv")
df = pd.merge(df, category_df[['商品番号', 'カテゴリー']], how='inner', on='商品番号')

category_value = df['カテゴリー'].value_counts()
category_value.plot(kind='bar')

plt.title('カテゴリーの出現頻度')
plt.xlabel('カテゴリー')
plt.ylabel('出現回数')

plt.show()

X = df.groupby('商品番号')['注文数'].describe()

print(X)
```
